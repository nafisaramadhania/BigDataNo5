```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv(r'C:\Users\ASUS\Downloads\Online Retail\Online Retail.csv')

data
```

```python
# Preprocessing
basket = data.groupby(['InvoiceNo', 'StockCode']).size().unstack(fill_value=0)
basket = basket > 0 

# Apriori
minsupp = 0.01
frequent_itemsets = apriori(basket, min_support=minsupp, use_colnames=True)

# Association Rules
minconf = 0.4 
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=minconf, num_itemsets=len(frequent_itemsets))

# Results
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].to_csv('rules_output.csv', index=False)

print("Data berhasil disimpan ke file 'rules_output.csv'")
```

```python
max_support_row = rules.loc[rules['support'].idxmax()]
max_confidence_row = rules.loc[rules['confidence'].idxmax()]
max_lift_row = rules.loc[rules['lift'].idxmax()]

print("Baris dengan support maksimum:")
print(max_support_row)

print("\nBaris dengan confidence maksimum:")
print(max_confidence_row)

print("\nBaris dengan lift maksimum:")
print(max_lift_row)
```
